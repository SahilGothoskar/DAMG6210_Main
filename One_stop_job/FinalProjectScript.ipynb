```python
#replace all the characters 
df['Salary'] = df['Salary'].str.replace('Estimated ','').str.replace(' a year','').str.replace('From','').str.replace('-','').str.replace('a month','').str.replace('an hour','').str.replace('Indeed','').str.replace('est','')
df['Salary'] = df['Salary'].str.replace('Fulltime','').str.replace('\n+1','').str.replace('Parttime','').str.replace('Up to','').str.replace('a year','').str.replace('$','').str.replace('.','').str.replace('Full-time','')
```

```python
#replace all the characters 
df['Salary1'] = df['Salary1'].str.replace(',','').str.replace('Up','').str.replace(' ','').str.replace('Contract','').str.replace('Internship','').str.replace('None','')
df['Salary1'] = df['Salary1'].str.replace('.','').str.replace('K','00').str.replace('Temporary','')
```

```python
#display Salary1
df['Salary1']
```

```python
#update the salary1
df['Salary1'] = salaries
```
