```python
#Importing libraries for dataframes and date time
import pandas as pd
from datetime import datetime, timedelta
```

```python
data_df = pd.read_csv('data.csv')  
```

```python
data_df
```

```python
data_df.drop(['company_id','today'], axis=1, inplace=True)
```

```python
data_df = data_df.rename(columns={'job_title':'job_position','company':'company_name',
                                          'post_date':'posted_date','job_summary':'job_description','employment_type':'job_type','id':'company_id','job_id':'id'})
```

```python
#cleaning post_date
data_df['posted_date'] = data_df['posted_date'].map(lambda x: x.lstrip('EmployerActive').lstrip('PostedPosted').lstrip('Hiring ongoing').lstrip('+'))
                                                   
                                                   
```

```python
posted_date = data_df['posted_date'].values.tolist()
```

```python
posted_date = [item.replace('day', 'days').replace('dayss','days') for item in posted_date]
```

```python
posted_date
```

```python
#changing format of post_date
post_dates_2 = []
for i in posted_date:
    if (i == ""):
        post_dates_2.append(" ")
    elif (i == "30+ days ago"):
        t = datetime.date.today() + datetime.timedelta(-30)
        t = t.strftime('%m/%d/%Y')
        post_dates_2.append(t)
    elif (i == "Todays" or i == "Just posted"):
        t = datetime.date.today()
        t = t.strftime('%m/%d/%Y')
        post_dates_2.append(t)
    else: 
        parsed_s = [i.split()[:2]]
        time_dict = dict((fmt,float(amount)) for amount,fmt in parsed_s)
        dt = datetime.timedelta(**time_dict)
        past_time = datetime.datetime.now() - dt
        t = past_time.date()
        t = t.strftime('%m/%d/%Y')
        post_dates_2.append(t)
```

```python
post_dates_2
```

```python
data_df['posted_date'] = post_dates_2
```

```python
data_df = data_df.iloc[: , 1:]
```

```python
data_df
```
