### Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Fetching data for State wise Tax Rate

```python
url='https://www.realized1031.com/capital-gains-tax-rate'
response = requests.get(url)
```

```python

soup = BeautifulSoup(response.text, 'html')
```

```python
print(soup)
```

### Extract Table data from HTML Fetch content

```python
# Obtain information from tag <table>
table1 = soup.find('table', attrs={"id": "tablesort"})
table1
```

### Getting Headers for Table

```python

headers = []

for i in table1.find_all("th"):
 title = i.text
 headers.append(title)

print(headers)
```

```python
mydata = pd.DataFrame(columns = headers)
```

### Adding Extracted Data to dataframe

```python
# Create a for loop to fill mydata
for j in table1.find_all('tr')[1:]:
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```
